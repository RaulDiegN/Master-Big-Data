# 1. Cargar la red social del club de kárate (1 punto)

Pista: buscad entre los generadores de NetworkX.


```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.karate_club_graph()
```

# 2. Dibujar la red social del club de kárate  (1 punto)

```python
nx.draw_circular(G, with_labels=True)
plt.show()
```

# 3. Datos básicos (1 punto)
1. Número de nodos y número de aristas.
2. Densidad.
3. Diámetro.
4. Grado máximo, mínimo y promedio.

```python
print("Número de nodos: " + str(len(G.nodes())) + " // Número de aristas " + str(len(G.edges())))
```

```python
print("densidad: %s" % nx.density(G))
```

```python
print("diámetro: %d" % nx.diameter(G))
```

```python
grados = list(dict(G.degree()).values())

print("El grado máximo es: " + str(max(grados)) + " // El grado mínimo es: " + str(min(grados)) + 
      " // El grado promedio es: " + str(sum(grados)/len(grados)))
```

# 4. Identifica al instructor y al administrador. (1 punto)

Son los dos nodos de mayor grado.

```python
max_1 = grados.index(max(grados))
grados.remove(max(grados))
max_2 = grados.index(max(grados))

print("El instructor y el administrador son los siguientes dos nodos: " 
      + str(max_1) + " y " + str(max_2))
```

# 5. Vuelve a dibujar el grafo pero, esta vez, resaltando los nodos del instructor y del administrador. (1 punto)

(Es suficiente con que tengan un color distinto, por ejemplo)

```python
color_map = ['red' if x == max_1 or x == max_2 else 'blue' for x in G.nodes()]
size_map = [470 if x == max_1 or x == max_2 else 250 for x in G.nodes()]

nx.draw_circular(G, with_labels=True, node_color=color_map, 
        node_size=size_map)
```

# 6. Calcula la centralidad de cercanía (closeness) y de intermediación (betweenness). (1 punto)

```python
nx.closeness_centrality(G)
```

```python
nx.betweenness_centrality(G)
```

# 7. Estudia el prestigio en la red mediante PageRank. (1 punto)

```python
nx.pagerank(G)
```

# 8. Obtén el coeficiente de agrupamiento medio y el número de triángulos completos (1 punto)

```python
print("El coeficiente de agrupamiento medio es " + str(nx.average_clustering(G)))
```

```python
print("El número de triangulos es " + str(sum(list(dict(nx.triangles(G)).values()))/3))
```

# 9. Obtén las comunidades según el algoritmo de Girvan-Newmann (1 punto)

```python
from networkx.algorithms import community

comp = community.girvan_newman(G)
tuple(sorted(c) for c in next(comp))
```
